В примере мы считываем числа 1, -3, 5, -6, -10, 13; в этот момент замечаем, что сумма этих чисел равна нулю и выводим сумму их квадратов, не обращая внимания на то, что остались ещё не прочитанные значения.

```python
final_list = []
final_sum = 0
a = int(input())
final_sum += a
while final_sum != 0:
    final_list.append(a)
    a = int(input())
    final_sum += a
final_list.append(a)
summa_kv = 0
for i in final_list:
    summa_kv += i**2
print(summa_kv)
```

```python
def modify_list(l):
    for i in range(len(l)-1,-1,-1):
        if l[i] % 2 != 0:
            l.pop(i)
        else:
            l[i] = int(l[i] / 2)
```
